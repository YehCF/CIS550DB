```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("sp500raw(original).csv")
df.head()
```

```python
df.rename(columns={"Unnamed: 0": "date"}, inplace=True)
```

```python
# get companies
companies = set()
for attr in df.columns:
    if "_" in attr:
        companies.add(attr.split("_")[0])
print(len(companies))
```

```python
out_df = {"date": [], "code": [], "high": [], "low": [], "close": [], "volume": []}
```

```python
for idx, row in df.iterrows():
    date_str = row["date"]
    attrs = ["low", "high", "close", "volume"]
    for code in list(companies):
        out_df["date"].append(date_str)
        out_df["code"].append(code)
        out_df["high"].append(row["{}_high".format(code, )])
        out_df["low"].append(row["{}_low".format(code, )])
        out_df["close"].append(row["{}_close".format(code, )])
        out_df["volume"].append(row["{}_volume".format(code, )])
```

```python
out_df = pd.DataFrame(out_df)
```

```python
# export
out_df.to_csv("sp500_price_test.csv", index=None)
```
